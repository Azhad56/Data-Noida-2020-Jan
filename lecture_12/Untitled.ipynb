```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-np.pi, np.pi, 100)
```

```python
y = (np.sin(X) + 2) * 1000
```

```python
plt.plot(X, y)
```

```python
plt.plot(X, np.log(y))
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_features=2, centers=3, random_state=42)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
x1 = np.linspace(-8, 8, 10)
plt.scatter(X[:, 0], X[:, 1], c=y)
for ((m1, m2), c) in zip(model.coef_, model.intercept_):
    x2 = -(m1*x1)/m2 - (c/m2)
    plt.plot(x1, x2)
plt.ylim(-10, 10)
    
```
