```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from tensorflow.keras.preprocessing.image import load_img

import numpy as np
import matplotlib.pyplot as plt
```

```python
img_pan = load_img("../datasets/panda.jpg", target_size=(300, 400))
img_mon = load_img("../datasets/mon.jpg", target_size=(300, 400))
```

```python
img_mon
```

```python
arr_pan = np.array(img_pan)
arr_mon = np.array(img_mon)
```

```python
arr_pan.shape, arr_mon.shape
```

```python
batch = np.array([arr_pan, arr_mon])
```

```python
batch.shape
```

```python
generator = ImageDataGenerator(rotation_range=10, horizontal_flip=True, shear_range=1.1)
```

```python
gen = generator.flow(batch, batch_size=10)
```

```python
for gen_img in gen.next():
    plt.figure()
    plt.imshow(gen_img.astype(int))
    plt.show()
```
